```python
import itertools as it
```

```python
lst = ['a','b','c']
```

```python
perms = it.permutations(lst, r=2)
perm_list = list(perms)
print(perm_list)
```

```python
perms = it.combinations(lst, r=2)
perm_list = list(perms)
print(perm_list)
```

```python
perms = it.combinations_with_replacement(lst, 2)
perm_list = list(perms)
print(perm_list)
```

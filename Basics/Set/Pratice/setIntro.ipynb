### Set is an unordered collection of hetrogenous unique elements. It is mutable and does not allow duplicates. It is defined by using curely braces {} or by using set() function.

```python
import jupyter

S1 = {1,2,3,4,5}
print(S1)
```

```python
S2 = set([1,2,3,4,5])
# jupyter notebook prints the set in ordered list
S2
```

### Set comprehension

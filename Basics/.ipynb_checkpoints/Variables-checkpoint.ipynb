# Integer Data Type

```python
import sys
x=15
```

```python
y=158976589886
```

```python
print(sys.getsizeof(x))
print(sys.getsizeof(y))
```

```python
id(x)
```

```python
x=16
id(x)
```

```python
abc = 15
print(id(abc))
abc=16
print(id(abc))
abc=17
print(id(abc))
```

#### Reuse of memory location
- as the id of 'x' while value 15 is same as the id of 'abc' while 15 and same goes for 16 as well.
- While the memory stores 15 has been de-referenced and x value has been initalized with 16 and referenced to new memory location, even after this also when we defined 'abc' with 15 it used the dereferenced memory lcoation only.

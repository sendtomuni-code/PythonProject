### Number Systems

```python
i = 10
j = 0b1010
k = 0o12
l = 0xA
m = 0xa
print(i, "-", j, "-", k, "-", l, "-", m)
```

```python
n = 0xa+5j
print(n)
```

```python
o = 0xa+0xaj
```

```python
p = 10
print(bin(p))
print(oct(p))
print(hex(p))
print(type(bin(p)), "-", type(oct(p)), "-", type(hex(p)))
```

```python
print(bin(True))
print(bin(False))
```

```python
print(bin(1.0))
```

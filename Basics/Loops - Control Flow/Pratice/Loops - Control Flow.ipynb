## Loop to get digits of a number

```python
n = 23456
while n > 0:
    r = n % 10
    print(r)
    n = n // 10
```

## Count Digits of a Number

```python
n = 23456
count = 0
while n > 0:
    n = n // 10
    count += 1
print(count)
```

## Sum of all digits of a number

```python
n = 23456
total = 0
while n > 0:
    r = n % 10
    total += r
    n = n // 10
print(total)
```

## Reverse a number

```python
n = int(input('Enter a number: '))
m = n
reverse = 0
while n > 0:
    r = n % 10
    reverse = reverse * 10 + r
    n = n // 10
print(reverse)
if m == reverse:
    print('The number is a palindrome.')
else:
    print('The number is not a palindrome.')
```

## Find Max and Min of a list

```python
n = int(input('Enter the number of elements: '))
numbers = []
for i in range(n):
    num = int(input('Enter a number: '))
    numbers.append(num)
Max = float('-inf')
Min = float('inf')
length = len(numbers)
i = 0
while i < length:
    if numbers[i] > Max:
        Max = numbers[i]
    if numbers[i] < Min:
        Min = numbers[i]
    i += 1
print('Maximum number:', Max)
print('Minimum number:', Min)
```

```python
n = int(input('Enter a number: '))
factorial = 1
for i in range(n, 1, -1):
    factorial *= i
print(factorial)
```

```python
n = int(input('Enter a number: '))
fibonacci = [0, 1]
for i in range(2, n):
    next_fib = fibonacci[i - 1] + fibonacci[i - 2]
    fibonacci.append(next_fib)
print(fibonacci[:n])
```

```python
# Showing Differences of Flow execution on continue, break and pass
for i in range(10):
    if i == 2:
        continue
    if i == 7:
        break
    if i == 4:
        pass
    print(i)
```

# 🚀 Operators and Expressions

when we use c = a + b, it is an expression. <br>
in this expression, a, b and c are operands and + is the operator. <br>

```python
a = 14
b = 4
```

## ➕, ➖, 💥 Addition, Substraction and Multiplication operator

```python
print(a, "+", b, "=", a+b)
print(a, "-", b, "=", a-b)
print(a, "*", b, "=", a*b)
```

## /, //, % Division, Floor Division and Modulus operator

"/" This operator stands for float division. It returns the quotient of the division as a float, even if both operands are integers. <br>
Unlike other programming language (like java, c), where "/" always gives the integer <br>
"//" This operator stands for floor division. It returns the quotient of the division as an integer, by discarding the fractional part. <br>
"%" This operator stands for modulus. It returns the remainder of the division. <br>

```python
print(a, "/", b, "=", a/b)
print(a, "//", b, "=", a//b)
print(a, "%", b, "=", a%b)
```

## **Exponentiation operator**
While we want to calculate the power of a number, we can use the exponentiation operator "**". <br>
It takes two operands, the base and the exponent, and returns the result of raising the base to the power of the exponent. <br>

```python
print(2, "**", 5, "=", 2**5)
```

## 🚀 Precedence of Operators

```python
print (2 + 3 * 4) # 14 First multiplication took place and then addition
print ((2 + 3) * 4) # 20 First addition took place due to parenthesis and then multiplication
```

```python
print(2 ** 3 ** 2) #
```

its 2 to the power of 3 to the power of 2. So, first 3 to the power of 2 is calculated and then 2 to the power of 9 is calculated. So, the result is 512. Because exponentiation operator is right to left associative.

```python
print( 2 + 5 - 3 * 12 / 2 % 3)
```

first multiplication is calculated, then division, then modulus. Because multiplication, division and modulus have same precedence and they are left to right associative. Then addition and subtraction is calculated. Because they have same precedence and they are left to right associative. So, the result is 4.0

```python
print(2 + (5 - (3 * 12) / (2 % (3 ** 2))))
```

first exponentiation is calculated, then modulus, then multiplication, then division, then subtraction and then addition. So, the result is -11.0

## 🚀 WAP to calculate the area of a Circle

```python
radius = 7
pi = 3.14
area = pi * (radius ** 2)
print("Area of Circle is:", area)
```

Python supports positive and negative indexing. The index of the first element is 0, the second element is 1, and so on. The index of the last element is -1, the second last element is -2, and so on.

```python
s1 = 'hello';

print(s1[0])  # Output: h
print(s1[1])  # Output: e
print(s1[-1]) # Output: o
print(s1[-2]) # Output: l
```

### String Slicing

Slicing is a way to extract a portion of a string. The syntax for slicing is `string[start:end:step]`, where `start` is the index to start the slice, `end` is the index to end the slice (not inclusive), and `step` is the step size (optional). From start to end, it has to be forward direction only, backward is not allowed. If you want to slice backward, you can use negative indexing.

```python
s1 = 'Hello World'
print("s1[:] -> " + s1[:])
print("s1[::] -> " + s1[::])

print("s1[::1] -> " + s1[::1])
print("s1[0::] -> " + s1[0::])
print("s1[:11:] -> " + s1[:11:])

s2 = s1[::-1]
print("s1[::-1] -> " + s2[::])

s3 = s1[11::]
print("s1[11::] -> " + s3[::])
s4 = s1[11:0:]
print("s1[11:0:] -> " + s4[::])
s5 = s1[11:-12:-1]
print("s1[11:-12:-1] -> " + s5[::])

s5 = s1[:0:]
print("s1[:0:1] -> " + s5[::])
```

When we SAY strip multiple characters, it means that we want to remove all the characters that are present in the given string from the beginning and end of the original string. It does not mean that we want to remove the exact sequence of characters from the original string. For example, if we have a string "abcde" and we want to strip "ab", it will remove both 'a' and 'b' from the beginning of the string, resulting in "cde". It will not remove the exact sequence "ab" from the string. It fetches from left and right side and discards the character specified, once it identifies a character that has not been specified to discard it stops, even if there are character after this character that has been specified to discard. It does not remove the characters from the middle of the string. Like in below even though character left parenthesis, right parenthsis, * is present in the middle of the string, it will not be removed, because it found '&' while scanning from right, which has not been instructed to strip.

```python
multi_char_padded_letter.strip('*~!@#$ ()*')
```

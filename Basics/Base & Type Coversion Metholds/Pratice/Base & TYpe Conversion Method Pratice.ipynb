## Type Conversion Methods

```python
a = 10
s1 = bin(a)
s2 = oct(a)
s3 = hex(a)
```

```python
print(a, "-", s1, "-", s2, "-", s3)
```

Type conversion methods only accept integers as input

```python
invalidBin = bin(10.5)
```

Type conversion methods accepts boolean values as input

```python
binBool = bin(True)
octBool = oct(False)
hexBool = hex(True)
print(binBool, "-", octBool, "-", hexBool)
```

# 📝 Type conversion methods


## 🚀 integer conversion method

int() - coverts a value to an integer. <br>
It can take float, boolean. <br>
<b> It cannot take complex numbers. </b>
it can take the string if the string is a valid integer. <br>
If the string is not a valid integer, it will raise a ValueError.
it can also take the different base numbers, but in this case it requires two arguments, the string and the base. <br>

```python
floatNumber = 10.5
print(type(floatNumber))
intNumber = int(floatNumber)
print(intNumber)
print(type(intNumber))
```

```python
print(int(True))
```

```python
print(int(complex(10,5)))
```

```python
print(int("1"))
print(int("10.5"))
```

```python
print(int("abc"))
```

```python
print(int("1010", 2))
print(int("12", 8))
print(int("A", 16))
```

## 🚀 float conversion method

float converts from integer, boolean. <br>
float caoonot covert from complex numbers. <br>
float can convert from string if the string is a valid float. <br>
there is no base conversion in float. <br>

ℹ️  Only integer has support to different base numbers

```python
i = 10
f = float(i)
print(type(i))
print(f)
print(type(f))
```

```python
print(float(True))
print(float(False))
```

```python
print(float(complex(10,5)))
```

```python
print(float("10.5"))
```

```python
print(float("abc"))
```

## 🚀 boolean conversion method

any values can be converted to boolean. <br>
By default for any value bool() function will return True, except for the following values: <br>
- False
- None
- 0

```python
print("0 ->", bool(0))
print("False ->", bool(False))
print("()) ->", bool())
print("None ->", bool(None))
print("0.0 ->", bool(0.0))
print("\"\" ->", bool(""))
print("0j ->", bool(0j))
```

```python
a = 10
print(type(a))
boolA = bool(a)
print(type(boolA))
print(boolA)
```

```python
print("1 ->", bool(1))
print("True ->", bool(True))
print("5.2 ->", bool(5.2))
print("abc ->", bool("abc"))
print("5j ->", bool(5j))
```

## 🚀 complex conversion method

Complex conversion also takes any value as input. <br>
By default, it will return a complex number with the real part as the input value and the imaginary part as 0. <br>
If the input value is a string, it should be a valid complex number in the form of "a+bj" or "a-bj". <br>
If the string is not a valid complex number, it will raise a ValueError. <br>
There is no base conversion in complex numbers. <br>

```python
i = 10
c = complex(i)
print(type(i))
print(c)
print(type(c))
```

```python
print(complex(True))
print(complex(False))
print(complex(10.5))
print(complex("5+6j"))
```

```python
print(complex("abc"))
```

## 🚀 String conversion method

All typed values can be converted to string. <br>
The string representation of the value will be returned. <br>
There is no base conversion in string. But different base numbers can be given to str function and it will show in correct format<br>

```python
i = 10
s = str(i)
print(type(i))
print(s)
print(type(s))
```

```python
print(str(True))
print(str(False))
print(str(10.5))
print(str(5+6j))
```

```python
print(str(bin(2)))
print(str(oct(2)))
print(str(hex(2)))
```

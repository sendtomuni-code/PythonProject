### Indexing and slicing

We cannot write with indexing on slicing on tuple, because tuple is immutable, but we can read the values from tuple using indexing and slicing.

```python
t10 = (1,2,3,4,5,6,7,8,9)
```

```python
t10[0]
```

```python
t10[4:7:1]
```

```python
t10[4:7]
```

```python
t10[::-1]
```

# Area of a trapezium

```python
a = int(input("Enter the shortest length of the trapezium: "))
b = int(input("Enter the longest length of the trapezium: "))
h = int(input("Enter the height of the trapezium: "))
area = (a+b) * h / 2
print("The area of the trapezium is: ", area)
```

# Area of a cuboid

```python
l = 5
b = 10
h = 7
area = 2 * ((l*b) + (b*h) + (h*l))
print("The area of the trapezium is: ", area)
```

# Solving a quadratic equation
ax2 + bx + c = 0 is a quadratic equation where a, b and c are constants and x is the variable. The solution of the quadratic equation can be found using the formula:
x = (-b ± √(b² - 4ac)) / (2a)


## Lets solve the quadratic equation 2x² + 3x - 5 = 0

```python
a=2
b=-7
c=3

x1 = (-b + math.sqrt(b**2 - 4*a*c)) / (2*a)
x2 = (-b - math.sqrt(b**2 - 4*a*c)) / (2*a)

print("The solutions of the quadratic equation are: ", x1, "and", x2)
```

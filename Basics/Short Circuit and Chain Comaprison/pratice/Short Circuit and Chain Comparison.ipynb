# Short Circuit Comparison

```python
a = 10
b = 5
c = 6

if a>b and a>c:
    print("a is greater than b and c")
# in and operator if first operator results true then second operator will be taken into consideration

if a<b and a<c:
    print("a is less than b and c")
# in and operator if first operator results false then second operator will be not be calculated as regardless of the second condition, and will result in false. So after first condition is false then it will not check for second condition and will directly result in false. This is called short-circuiting in python.
```

# Bit-wise Operators

```python
print(f'10 & 13 is: {10 & 13}')
# 10 - 1010
# 13 - 1101
#    - 1000 - 8
print('  ' + format(10, 'b'))
print('& ' + format(13, 'b'))
print('------')
print('  ' + format(8, 'b'))
```

```python
print(f'10 | 13 is: {10 | 13}')
# 10 - 1010
# 13 - 1101
#    - 1111 - 15
print('  ' + format(10, 'b'))
print('| ' + format(13, 'b'))
print('------')
print('  ' + format(15, 'b'))
```

```python
print(f'10 ^ 13 is: {10 ^ 13}')
# 10 - 1010
# 13 - 1101
#    - 0111 - 7
print('  ' + format(10, 'b'))
print('& ' + format(13, 'b'))
print('------')
print('  0' + format(7, 'b'))
```

```python
print(f'~ of 10 is: {~10}')
# 10 - 1010
#    - 10100 --> 10101  - 5
print('  ' + format(10, 'b'))
print('& ' + format(13, 'b'))
print('------')
print('  0' + format(7, 'b'))
```
